```python
import numpy as np
import faiss
from sklearn.metrics import roc_auc_score, average_precision_score
from tqdm import tqdm
import pandas as pd

k_value = 1

train_set = np.load('/content/features/real/features.npy', allow_pickle=True).astype(np.float32)
index = faiss.IndexFlatL2(train_set.shape[1])
index.add(train_set)

df_train = pd.DataFrame(columns=["filename", "feature1", "feature2", "feature3", "feature4", "feature5", "label"])

for folderID in os.listdir('/content/features/fake-250'):
  test_set = np.load('/content/features/fake-250/'+folderID+"/features.npy", allow_pickle=True).astype(np.float32)
  features, _ = index.search(test_set, k_value)

  new_row = {"filename": "Celeb-Youtube-fake-sample/"+folderID+".mp4", "feature1": float(features[0]), "feature2": float(features[1]),
               "feature3": float(features[2]), "feature4": float(features[3]), "feature5": float(features[4]), "label": 1}
  df_train.loc[len(df_train.index)] = new_row

for folderID in os.listdir('/content/features/real-125-celeb'):
  try:
    test_set = np.load('/content/features/real-125-celeb/'+folderID+"/features.npy", allow_pickle=True).astype(np.float32)
    features, _ = index.search(test_set, k_value)

    new_row = {"filename": "Celeb-real/"+folderID+".mp4", "feature1": float(features[0]), "feature2": float(features[1]),
                 "feature3": float(features[2]), "feature4": float(features[3]), "feature5": float(features[4]), "label": 0}
    df_train.loc[len(df_train.index)] = new_row
  except IndexError:
    print(features)
    pass

for folderID in os.listdir('/content/features/real-125-yt'):
  try:
    test_set = np.load('/content/features/real-125-yt/'+folderID+"/features.npy", allow_pickle=True).astype(np.float32)
    features, _ = index.search(test_set, k_value)

    new_row = {"filename": "YouTube-real/"+folderID+".mp4", "feature1": float(features[0]), "feature2": float(features[1]),
                 "feature3": float(features[2]), "feature4": float(features[3]), "feature5": float(features[4]), "label": 0}
    df_train.loc[len(df_train.index)] = new_row
  except IndexError:
    pass

df_train
```

```python
df  = df_train
feature_cols = ["feature"+str(k) for k in range(1,6)]
df[feature_cols] = df[feature_cols].replace(0, np.nan)

# Step 3: Calculate the mean of these columns, ignoring NaN values
df['mean_column'] = df[feature_cols].apply(lambda x: x.mean(), axis=1)

# Step 4: Assign these means to the blank column and drop rows where all five features were zero
df['blank_column'] = df['mean_column']
df = df.dropna(subset=feature_cols, how='all')
```

# Train linear regressor

```python
import tensorflow as tf
from sklearn.linear_model import LinearRegression
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping, LearningRateScheduler

model = LinearRegression()
```

```python
import numpy as np
from sklearn.model_selection import train_test_split

# inputs and labels
X = df2.drop(columns = ['feature2', 'feature3', 'filename'], axis=1).values  # put all extra columns here
y = df2['feature2'].values

X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.01, random_state=42)

model.fit(X_train, y_train)
```

```python
df2['k'] = model.predict(df2.drop(columns = ['pred', 'feature2', 'feature3', 'filename'], axis=1).values)
```

```python
def breakpoint(x):
  if x>0.625:
    return 1
  else:
    return 0

df2['pred'] = df2['k'].apply(breakpoint)
df2
```

```python
from sklearn.metrics import f1_score

# Assuming df is your DataFrame and 'col1', 'col2' are the names of the columns you want to compare
df2['pred'] = df2['pred'].astype(int) # Convert to integer type
df2['feature2'] = df2['feature2'].astype(int) # Convert to integer type

y_true = df2['pred'].values
y_pred = df2['feature2'].values

f1 = f1_score(y_true, y_pred, average='binary')
print(f"F1 Score: {f1}")
```
