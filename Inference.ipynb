```python
! rm -rf /content/frames_out
! mkdir /content/frames_out /content/frames_out/Celeb-real /content/frames_out/Celeb-Youtube-fake-sample /content/frames_out/YouTube-real /content/frames_out/test
```

```python
import cv2
import numpy as np
import os
from tqdm import tqdm

numframes = 5
inpaths = ['/content/deep-fake-dataset/Celeb-real', '/content/deep-fake-dataset/YouTube-real', '/content/deep-fake-dataset/test', '/content/deep-fake-dataset/Celeb-Youtube-fake-sample']
outpaths = ['/content/frames_out/Celeb-real', '/content/frames_out/YouTube-real', '/content/frames_out/test', '/content/frames_out/Celeb-Youtube-fake-sample']

for idx in range(0,4):
  to_iterate = list(os.walk(inpaths[idx]))
  for root, dirs, files in tqdm(to_iterate, total=len(to_iterate)):
      for file in files:
          if file.endswith('.mp4'):
              image_path = os.path.join(root, file)
              vidcap = cv2.VideoCapture(image_path)
              frame_count = int(vidcap.get(cv2.CAP_PROP_FRAME_COUNT))

              # Fixed number of frames (same interval between frames)
              frame_idxs = np.linspace(0, frame_count - 1, numframes, endpoint=True, dtype=int)

              out_path = image_path.replace(inpaths[idx], outpaths[idx])[:-4] + '-' # Cut .mp4 suffix
              if not os.path.exists(os.path.dirname(out_path)):
                  os.makedirs(os.path.dirname(out_path))
              success, image = vidcap.read()
              count = 0
              while success:
                  if count not in frame_idxs:
                      success, image = vidcap.read()
                      count += 1
                      continue
                  cur_out_path = (out_path+'frame%d.jpg' % count)
                  cv2.imwrite(cur_out_path, image)  # save frame as JPEG file
                  success, image = vidcap.read()
                  count += 1
              vidcap.release()
```

# Detect faces and crop images for better accuracy

```python
%cd /content/FaceX-Zoo/face_sdk

! rm -rf /content/aligned_out
! mkdir /content/aligned_out /content/aligned_out/Celeb-real /content/aligned_out/Celeb-Youtube-fake-sample /content/aligned_out/YouTube-real /content/aligned_out/test

from core.image_cropper.arcface_cropper.FaceRecImageCropper import FaceRecImageCropper
from core.model_handler.face_alignment.FaceAlignModelHandler import FaceAlignModelHandler
from core.model_handler.face_detection.FaceDetModelHandler import FaceDetModelHandler
from core.model_loader.face_alignment.FaceAlignModelLoader import FaceAlignModelLoader
from core.model_loader.face_detection.FaceDetModelLoader import FaceDetModelLoader
import math
import sys, os
import cv2
import yaml
import numpy as np
from tqdm import tqdm

def distance2center(x1, y1, x2, y2, image):
    im_cx = int(image.shape[1] / 2)
    im_cy = int(image.shape[0] / 2)
    cx = ((x2 + x1) / 2).astype(int)
    cy = ((y2 + y1) / 2).astype(int)
    return math.sqrt(math.pow(im_cx - cx, 2) + math.pow(im_cy - cy, 2))


def Filter2centerBox(boundingBoxes, frame):
    min_distance = 999999999
    min_idx = -1
    for i, det in enumerate(boundingBoxes):
        distance = distance2center(det[0], det[1], det[2], det[3], frame)
        # cv2.rectangle(frame, (int(det[0]), int(det[1])), (int(det[2]), int(det[3])), (2, 255, 0), 1)
        # cv2.imshow("x", frame)
        # cv2.waitKey(0)
        if distance < min_distance:
            min_idx = i
            min_distance = distance
    return np.array([boundingBoxes[min_idx]])


def AlignedOneImageUsingFaceXAlignment(input_root, out_root, image_path):
    # print("aligning {}".format(image_path))
    try:
        image = cv2.imread(image_path, cv2.IMREAD_COLOR)
        input_height, input_width, _ = image.shape
    except:
        return
    dets = faceDetModelHandler.inference_on_image(image)
    if len(dets) > 0:
        dets = Filter2centerBox(dets, image)
        for i, det in enumerate(dets):
            assert (i != 1)  # only one face in picture
            landmarks = faceAlignModelHandler.inference_on_image(image, det)
            # print(landmarks.shape)
            cropped_image = face_cropper.crop_image_by_mat(image, landmarks.reshape(-1))
            out_path = image_path.replace(input_root, out_root)
            if os.path.exists(os.path.dirname(out_path)) is False:
                os.makedirs(os.path.dirname(out_path))
            cv2.imwrite(out_path, cropped_image)
    else:
        out_path = image_path.replace(input_root, out_root)
        cv2.imwrite(out_path, image)

inpaths = ['/content/frames_out/Celeb-real', '/content/frames_out/YouTube-real', '/content/frames_out/test', '/content/frames_out/Celeb-Youtube-fake-sample']
outpaths = ['/content/aligned_out/Celeb-real', '/content/aligned_out/YouTube-real', '/content/aligned_out/test', '/content/aligned_out/Celeb-Youtube-fake-sample']

with open('./config/model_conf.yaml') as f:
      model_conf = yaml.load(f, yaml.FullLoader)

model_path = 'models'

for idx in range(0,4):
  # detect init
  scene = 'non-mask'
  model_category = 'face_detection'
  model_name = model_conf[scene][model_category]
  faceDetModelLoader = FaceDetModelLoader(model_path, model_category, model_name)
  modelDet, cfgDet = faceDetModelLoader.load_model()
  faceDetModelHandler = FaceDetModelHandler(modelDet, 'cuda:0', cfgDet)

  # alignment init
  model_category = 'face_alignment'
  model_name = model_conf[scene][model_category]
  faceAlignModelLoader = FaceAlignModelLoader(model_path, model_category, model_name)
  modelAli, cfgAli = faceAlignModelLoader.load_model()
  faceAlignModelHandler = FaceAlignModelHandler(modelAli, 'cuda:0', cfgAli)

  # face cropper
  face_cropper = FaceRecImageCropper()

  to_iterate = list(os.walk(inpaths[idx]))
  for root, dirs, files in tqdm(to_iterate, total=len(to_iterate)):
      for file in files:
          AlignedOneImageUsingFaceXAlignment(inpaths[idx], outpaths[idx], os.path.join(root, file))
```

```python
import os
import shutil

# Source directory
src_dir = "/content/aligned_out/test"

# Destination directory root
dst_dir_root = "/content/features/test"

# Iterate through files in source directory
for filename in os.listdir(src_dir):
    # Extract ID and frame number from filename
    ID, frame_num = filename.split("-")

    # Construct destination directory for the current ID
    dst_dir = os.path.join(dst_dir_root, ID)

    # Create the destination directory if it doesn't exist
    if not os.path.exists(dst_dir):
        os.makedirs(dst_dir)

    # Construct source and destination paths
    src_path = os.path.join(src_dir, filename)
    dst_path = os.path.join(dst_dir, frame_num)

    # Move the file
    shutil.copy(src_path, dst_path)
```

```python
import numpy as np
import os
import faiss
from tqdm import tqdm
import pandas as pd

k_value = 1

train_set = np.load('/content/features/real/features.npy', allow_pickle=True).astype(np.float32)
index = faiss.IndexFlatL2(train_set.shape[1])
index.add(train_set)

df_test = pd.DataFrame(columns=["filename", "feature1", "feature2", "feature3", "feature4", "feature5"])

for folderID in os.listdir('/content/features/test'):
  test_set = np.load('/content/features/test/'+folderID+"/features.npy", allow_pickle=True).astype(np.float32)
  features, _ = index.search(test_set, k_value)

  new_row = {"filename": folderID+".mp4", "feature1": float(features[0]), "feature2": float(features[1]),
               "feature3": float(features[2]), "feature4": float(features[3]), "feature5": float(features[4])}
  df_test.loc[len(df_test.index)] = new_row

df_test
```

```python
df  = df_test
feature_cols = ["feature"+str(k) for k in range(1,6)]
df[feature_cols] = df[feature_cols].replace(0, np.nan)

# Step 3: Calculate the mean of these columns, ignoring NaN values
df['mean_column'] = df[feature_cols].apply(lambda x: x.mean(), axis=1)

# Step 4: Assign these means to the blank column and drop rows where all five features were zero
df['blank_column'] = df['mean_column']
df = df.dropna(subset=feature_cols, how='all')
```
